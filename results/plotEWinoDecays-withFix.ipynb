# Plot results using SLHA data

```python
import warnings
warnings.filterwarnings("ignore", message="numpy.dtype size changed")
import numpy as np
import glob,imp,os
import pyslha
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from collections import OrderedDict
import seaborn as sns
import xml.etree.ElementTree as ET
from scipy.interpolate import interp1d

# plt.style.use('fivethirtyeight') #Set style
# mpl.rcParams.update({'figure.figsize' : (15,10)})  #Set general plotting options

#Define plotting style:
sns.set() #Set style
sns.set_style('ticks',{'font.family':'Times New Roman', 'font.serif':'Times New Roman'})
sns.set_context('paper', font_scale=1.8)
cm = plt.cm.get_cmap('RdYlBu')
sns.set_palette(sns.color_palette("Paired"))
```

```python
particlesDict = {}
# with open("particles.xml") as f:
with open("particlesLatex.xml") as f:
    xmlStr = f.read()
    xmlStr = "<root>\n" + xmlStr[:] + "</root>" 
root = ET.fromstring(xmlStr)
for particle in root:
    particleData  = particle.attrib
    pid = int(particleData['id'])
    particlesDict[pid] = {}
    for attr,val in particle.attrib.items():
        try:
            particlesDict[pid][attr] = eval(val) #Try to evaluate attribute
        except:
            particlesDict[pid][attr] = val
```

### Load results:

```python
data = {}
for f in glob.glob('../EWino/data/slha_scan3new/*slha'):
    data[os.path.basename(f)] = pyslha.readSLHAFile(f)
```

#### Get all decay channels

```python
decaysDictC1 = {}
decaysDictN2 = {}
for d in data.values():
    for dec in d.decays[1000024].decays:
        fs = ','.join(sorted([particlesDict[abs(pid)]['name'] for pid in dec.ids]))
        if not fs in decaysDictC1:
            decaysDictC1[fs] = []    
            
    for dec in d.decays[1000023].decays:
        fs = ','.join(sorted([particlesDict[abs(pid)]['name'] for pid in dec.ids]))
        if not fs in decaysDictN2:
            decaysDictN2[fs] = []                
```

#### Get info

```python
ctau = []
tau = []
mC1 = []
mN2 = []
mN1 = []
M1 = []
M2 = []
tanb = []
mu = []
wC1 = []
for d in data.values():
    M1.append(d.blocks['EXTPAR'][1])
    M2.append(d.blocks['EXTPAR'][2])
    mu.append(d.blocks['EXTPAR'][23])
    tanb.append(d.blocks['EXTPAR'][25])
    wC1.append(d.decays[1000024].totalwidth)
    ctau.append(1.967e-16/d.decays[1000024].totalwidth)
    tau.append(6.582e-25/d.decays[1000024].totalwidth)
    mC1.append(d.blocks['MASS'][1000024])
    mN2.append(d.blocks['MASS'][1000023])
    mN1.append(d.blocks['MASS'][1000022])
    dDict = dict(zip(decaysDictC1.keys(),[0]*len(decaysDictC1)))
    for dec in d.decays[1000024].decays:
        fs = ','.join(sorted([particlesDict[abs(pid)]['name'] for pid in dec.ids]))        
        dDict[fs] = dec.br
    for fs in dDict:
        decaysDictC1[fs].append(dDict[fs])

    dDict = dict(zip(decaysDictN2.keys(),[0]*len(decaysDictN2)))
    for dec in d.decays[1000023].decays:
        fs = ','.join(sorted([particlesDict[abs(pid)]['name'] for pid in dec.ids]))        
        dDict[fs] = dec.br
    for fs in dDict:
        decaysDictN2[fs].append(dDict[fs])
        
        

mC1 = np.array(mC1)
mN1 = np.array(mN1)
ctau = np.array(ctau)
tau = np.array(tau)
wC1 = np.array(wC1)
M1 = np.array(M1)
M2 = np.array(M2)
mu = np.array(mu)
tanb = np.array(tanb)
for fs,v in decaysDictC1.items():
    decaysDictC1[fs] = np.array(v)
for fs,v in decaysDictN2.items():
    decaysDictN2[fs] = np.array(v)    
```

```python
### Plot from Dirac Gaugino paper (2007.08498 - Fig.1a)
dG = np.genfromtxt('../EWino/charginoLifetime.csv',delimiter=',',names=True)
```

```python
fig = plt.figure(figsize=(7,5))
plt.plot((mC1-mN1)[(mC1-mN1).argsort()],tau[(mC1-mN1).argsort()],linewidth=3,label='SoftSUSY')

# plt.scatter((mC1-mN1)[(mC1-mN1).argsort()],tau[(mC1-mN1).argsort()],linewidth=3,label='SoftSUSY')
plt.plot(dG['x'],dG['Curve1'],linewidth=3,label='Dirac Gaugino',c='red',alpha=0.6,linestyle='--')
plt.yscale('log')
plt.xscale('log')
plt.xlabel(r'$m_{C1}-m_{N1}$ (GeV)')
plt.ylabel(r'$\tau$ (s)')
plt.xlim(0.1,100)
plt.legend()
# plt.savefig('winoLifetime.png')
plt.grid()
plt.show()
```

```python
### Plot from Dirac Gaugino paper (2007.08498 - Fig.1a)
BR1pion = np.genfromtxt('../EWino/charginoBRs.csv',delimiter=',',names=True,skip_footer=42)
BR2pion = np.genfromtxt('../EWino/charginoBRs.csv',delimiter=',',names=True,skip_footer=25,skip_header=22)
```

```python
fig = plt.figure(figsize=(10,6))
for fs in decaysDictC1:
    plt.plot((mC1-mN1)[(mC1-mN1).argsort()],decaysDictC1[fs][(mC1-mN1).argsort()],label=fs,linewidth=3)
    plt.scatter((mC1-mN1)[(mC1-mN1).argsort()],decaysDictC1[fs][(mC1-mN1).argsort()])

plt.plot(BR1pion['x'],BR1pion['Curve2'],label=r'DG (1 $\pi$)',alpha=0.5,linewidth=3,linestyle='--')    
plt.plot(BR2pion['x'],BR2pion['Curve3'],label=r'DG (2 $\pi$)',alpha=0.5,linewidth=3,linestyle='--')    
    
plt.vlines(x=1.5,ymin=0,ymax=1,linewidth=1,alpha=0.6,color='gray',label=r'$\Lambda_{had}$')
plt.xlabel(r'$m_{\tilde{\chi}_1^+}-m_{\tilde{\chi}_1^0}$ (GeV)')
plt.ylabel(r'$BR$')
plt.legend(loc=(1.05,0.1))
plt.xlim(0.1,100)
plt.xscale('log')
# plt.xticks([0.2,0.5,1,1.5,2,5,80])
plt.grid()
plt.title(r'$M_2 = %1.1f, \mu = %1.1f, \tan\beta = %1.1f$' %(M2[0],mu[0],tanb[0]),fontsize=15)
plt.tight_layout()
# plt.savefig('winoDecays.png')
plt.show()
```

### Compare partial widths:

```python
### Dirac gaugino values
dGwidthF = interp1d(dG['x'],6.582e-25/dG['Curve1'])
dGwidth1pi = dGwidthF(BR1pion['x'])*BR1pion['Curve2']
dGwidth2pi = dGwidthF(BR2pion['x'])*BR2pion['Curve3']

### SoftSUSY values
width1pi = decaysDictC1['$\\pi^+$,$\\tilde{\\chi}_{1}^{0}$']*wC1
width2pi = decaysDictC1['$\\pi^+$,$\\pi^0$,$\\tilde{\\chi}_{1}^{0}$']*wC1
```

```python
fig = plt.figure(figsize=(10,6))
plt.plot((mC1-mN1)[(mC1-mN1).argsort()],width1pi[(mC1-mN1).argsort()],label='1 pion',linewidth=3)
plt.plot((mC1-mN1)[(mC1-mN1).argsort()],width2pi[(mC1-mN1).argsort()],label='2 pion',linewidth=3)

plt.plot(BR1pion['x'],dGwidth1pi,label='1 pion (DG)',linewidth=3,linestyle='--')
plt.plot(BR2pion['x'],dGwidth2pi,label='2 pion (DG)',linewidth=3,linestyle='--')

plt.vlines(x=1.5,ymin=0,ymax=1,linewidth=1,alpha=0.6,color='gray',label=r'$\Lambda_{had}$',linestyle='-.')
plt.xlabel(r'$m_{\tilde{\chi}_1^+}-m_{\tilde{\chi}_1^0}$ (GeV)')
plt.ylabel(r'$\Gamma_{\tilde{\chi}_1^+ \to X}$')
plt.legend()
plt.xlim(0.1,10)
plt.ylim(1e-18,1e-12)
# plt.ylim(1e-13,5e-13)
plt.xscale('log')
plt.yscale('log')
# plt.xticks([0.2,0.5,1,1.5,2,5,80])
plt.grid()
plt.title(r'$M_2 = %1.1f, \mu = %1.1f, \tan\beta = %1.1f$' %(M2[0],mu[0],tanb[0]),fontsize=15)
plt.tight_layout()
# plt.savefig('winoDecays.png')
plt.show()
```
