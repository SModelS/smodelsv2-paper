```python
import os
import xslha
from importlib import import_module
import pandas as pd
import sys
import numpy as np

sys.path.append('/home/alguero/Work/dataScalar')

# Fetching the slha and python outputs filenames
points = os.listdir("/home/alguero/Work/dataScalar/scanpoints")
shorts = [os.path.splitext(point)[0] for point in points]
pys = ["output."+sh for sh in shorts]
fpys = ["foutput."+sh for sh in shorts]
slhas = [os.path.join("/home/alguero/Work/dataScalar/scanpoints", slha) for slha in points]
```

```python
# Fetching particle masses and widths in slhas and reading python outputs
data = {'mA0':[], 'mHc':[], 'mH0':[], 'deltaM':[], 'dHc':[], 'rmax':[], 'ana':[], 'rmax2':[], 'ana2':[], 'rmaxDT':[], 'rmaxHSCP':[],'topo':[], 'bestTx':[], 'pions':[], 'slha':[]}
for slha, py in zip(slhas, pys):
    try:
        f = xslha.read(slha)
    except UnicodeDecodeError:
        print("could not read", slha)
    data['slha'].append(os.path.basename(slha))
    if not "MASS" in f.blocks:
        print("no MASS block in", slha)
        continue
    mA0 = f.Value("MASS", [36])
    mHc = f.Value("MASS", [37])
    mH0 = f.Value("MASS", [35])
    deltaM = mHc - mH0
    brHc = f.br[37]
    # Computing BR(A0->H+)
#     brA0 = f.br[36]
#     A0Hc = 0
#     for b in brA0:
#         if 37 in b:
#             A0Hc += 
    # Is H+ decaying to pions or not
    pions = False
    for b in brHc:
        if 211 in b:
            pions = True
            break
    data['pions'].append(pions)
    # Converting widths to ctau in meters
    dHc = 1/f.Value("WIDTH", 37)*6.582E-25*299792458
    data['mA0'].append(mA0)
    data['mHc'].append(mHc)
    data['mH0'].append(mH0)
    data['deltaM'].append(deltaM)
    data['dHc'].append(dHc)
    try:
        dic = import_module(py).smodelsOutput
    except ModuleNotFoundError:
        print("could not load", py)
    res = dic['ExptRes']
    rmax = res[0]['r']
    data['rmax'].append(rmax) 
    data['ana'].append(res[0]['AnalysisID'])
    data['topo'].append(res[0]['TxNames'])
    # Looking for the "best" txname
    weights = res[0]['TxNames weights (fb)']
    bestW = 0
    for tx in weights:
        if weights[tx] > bestW:
            bestW =  weights[tx]
            bestTx = tx
    data['bestTx'].append(bestTx)
    rmax2 = res[1]['r']
    data['rmax2'].append(rmax2) 
    data['ana2'].append(res[1]['AnalysisID'])
    # computing rmaxDT
    rmaxDT = 0
    for r in res:
        if 'TDT' in ','.join(r['TxNames']):
            rmaxDT = r['r']
            break
    data['rmaxDT'].append(rmaxDT)
    rmaxHSCP = 0
    for r in res:
        if 'THSCP' in ','.join(r['TxNames']):
            if rmaxHSCP < r['r']:
                rmaxHSCP = r['r']
    data['rmaxHSCP'].append(rmaxHSCP)
df = pd.DataFrame(data)
df.to_csv("scalarData.csv", index=False)
```

```python
# Just testing pylha
points = os.listdir("/home/alguero/Work/dataScalar/mysimplescan")
file = os.path.join("/home/alguero/Work/dataScalar/mysimplescan", points[1])
width=10.
with open(file) as f:
    fx = f.readlines()
with open('testFile.slha', 'w') as out:
    for l in fx:
        if 'DECAY 37' in l:
            out.write('DECAY 37  '+'{:.6E}'.format(width)+'  # ~H+\n')
        else:
            out.write(l)
```

```python
# Adding the scan over the H+ width
points = os.listdir("/home/alguero/Work/dataScalar/mysimplescan")
logctaus = np.arange(-2, 0.5, 0.2)
for p in points:
    file = os.path.join("/home/alguero/Work/dataScalar/mysimplescan", p)
    name = os.path.splitext(p)[0]
    with open(file) as f:
        lines = f.readlines()
    for logctau in logctaus:
        if abs(logctau) < 1e-10:
            logctau = 0
        newname = name+'logctau'+'{0:.2g}'.format(logctau*10)+'_.slha'
        newname = os.path.join("/home/alguero/Work/dataScalar/idm-simplescanWidth", newname)
        ctau = 10**logctau
        width = 1/ctau*6.582E-25*299792458
        with open(newname, 'w') as f:
            for line in lines:
                if 'DECAY 37' in line:
                    out.write('DECAY 37  '+'{:.6E}'.format(width)+'  # ~H+\n')
                else:
                    out.write(line)
```
