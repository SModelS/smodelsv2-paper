## Converting output files into a single csv file

```python
import os
import xslha
from importlib import import_module
import pandas as pd
import sys
import numpy as np

sys.path.append('/home/alguero/Work/dataScalar')

# Fetching the slha and python outputs filenames
points = os.listdir("/home/alguero/Work/dataScalar/scanpoints")
shorts = [os.path.splitext(point)[0] for point in points]
pys = ["output."+sh for sh in shorts]
fpys = ["foutput."+sh for sh in shorts]
slhas = [os.path.join("/home/alguero/Work/dataScalar/scanpoints", slha) for slha in points]
```

```python
# Fetching particle masses and widths in slhas and reading python outputs
data = {'mA0':[], 'mHc':[], 'mH0':[], 'deltaM':[], 'dHc':[], 'dataType':[], 'rmax':[], 'ana':[], 'rmax2':[], 'ana2':[], 'rmaxDT':[], 'rmaxHSCP':[],'topo':[], 'bestTx':[], 'pions':[], 'slha':[]}
for slha, py in zip(slhas, pys):
    try:
        f = xslha.read(slha)
    except UnicodeDecodeError:
        print("could not read", slha)
    data['slha'].append(os.path.basename(slha))
    if not "MASS" in f.blocks:
        print("no MASS block in", slha)
        continue
    mA0 = f.Value("MASS", [36])
    mHc = f.Value("MASS", [37])
    mH0 = f.Value("MASS", [35])
    deltaM = mHc - mH0
    brHc = f.br[37]
    # Is H+ decaying to pions or not
    pions = False
    for b in brHc:
        if 211 in b:
            pions = True
            break
    data['pions'].append(pions)
    # Converting widths to ctau in meters
    dHc = 1/f.Value("WIDTH", 37)*6.582E-25*299792458
    data['mA0'].append(mA0)
    data['mHc'].append(mHc)
    data['mH0'].append(mH0)
    data['deltaM'].append(deltaM)
    data['dHc'].append(dHc)
    try:
        dic = import_module(py).smodelsOutput
    except ModuleNotFoundError:
        print("could not load", py)
    res = dic['ExptRes']
    data['dataType'].append(res[0]['dataType'])
    rmax = res[0]['r']
    data['rmax'].append(rmax) 
    data['ana'].append(res[0]['AnalysisID'])
    data['topo'].append(res[0]['TxNames'])
    # Looking for the "best" txname
    weights = res[0]['TxNames weights (fb)']
    bestW = 0
    for tx in weights:
        if weights[tx] > bestW:
            bestW =  weights[tx]
            bestTx = tx
    data['bestTx'].append(bestTx)
    rmax2 = res[1]['r']
    data['rmax2'].append(rmax2) 
    data['ana2'].append(res[1]['AnalysisID'])
    # computing rmaxDT
    rmaxDT = 0
    for r in res:
        if 'TDT' in ','.join(r['TxNames']):
            rmaxDT = r['r']
            break
    data['rmaxDT'].append(rmaxDT)
    rmaxHSCP = 0
    for r in res:
        if 'THSCP' in ','.join(r['TxNames']):
            if rmaxHSCP < r['r']:
                rmaxHSCP = r['r']
    data['rmaxHSCP'].append(rmaxHSCP)
df = pd.DataFrame(data)
df.to_csv("scalarData.csv", index=False)
```
